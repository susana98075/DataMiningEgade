```python
x={'company':['Ford','Ford','VW','Audi','Audi','Ford'],
     'price':[2,3,1,2,3,1],'weight':[4,6,2,2,4,3],'lenght':[6,7,3,6,8,np.nan],'made':['China','Mexico','Mexico','Corea','Uk','Uk']}
```

```python
df=pd.DataFrame(data=x);df
```

```python
df.isnull().values.any()
```

```python
df['weight'].isnull().values.any()
```

```python
df['lenght'].isnull().values.any()
```

```python
df['lenght'].fillna(value=df.lenght.mean(),inplace=True);df
```

- What is the average price of audi cars mande in china?

```python
pd.pivot_table(df,index='company',
               aggfunc = ['sum','mean'],
               values='price',
               columns='made',
               margins=True,
               margins_name='Grand Total'
              )
```

```python
pd.pivot_table(df,index='company',
               aggfunc = {'price':np.mean,'weight':np.sum},
               values=['price','weight'],
               columns='lenght',
               margins=True,
               margins_name='Grand Total'
              )
```

```python
pd.pivot_table(df,index='company',
               aggfunc = {'price':np.mean,'weight':np.sum},
               values=['price','weight'],
               columns='made',
               margins=True,
               margins_name='Grand Total'
              )
```
