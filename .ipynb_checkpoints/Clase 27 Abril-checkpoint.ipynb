### loc vs.iloc

- loc 
    - df.loc[rowstar:rowend,columnstar:columnend] you need to use id names
    
- iloc 
    - df.iloc[rowstar:rowend,columnstar:columnend] you need to use id indices

```python
df[[1980]]
```

```python
df['Total']=df.iloc[:,3:].sum(axis=1);df.head(4)
```

## CREATE DATA FRAME

```python
df1=pd.DataFrame({'a':[3,4,5,6,7,8],'b':['Three','Four','Five','Six','Seven','Eight']});df1
```

```python
df2=pd.DataFrame({'c':[9,10,11,12,13,15],'d':['Nine','Ten','Eleven','Twelve','Thirteen','Fourteen']});df2
```

```python
df3=pd.concat([df1,df2],axis=1);df3
```

```python
df4=pd.concat([df1,df2],axis=0);df4
```

```python
df4.dropna(axis=0,inplace=True);df4
```
