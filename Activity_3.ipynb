```python
df=pd.read_csv('datasets/antropometria-dataset.csv')
df.head(4)
```

```python
df.columns
```

```python
pivot1=pd.pivot_table(df,index=['desc_ent'],
               aggfunc = 'mean',
               values=['peso']
              )
pivot1
```

```python
pivot1=pd.pivot_table(df,index=['desc_ent'],
               aggfunc = 'median',
               values=['peso']
              )
pivot1
```

```python
pivot2=pd.pivot_table(df,index=['desc_ent'],
               aggfunc = 'mean',
               values=['peso'],
               columns=['sexo'],
               sort=True
              )

pivot2
```

```python
pivot2.iloc[:,0].min()
```
