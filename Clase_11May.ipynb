```python
df=pd.read_csv('datasets/Cartwheeldata.csv')
df.head(4)
```

```python
sns.lmplot(x='Wingspan',y='CWDistance',data=df,fit_reg=False,hue='Gender')
plt.title('Scatter Plot Of CWDistance vs. Wingspan');
```

```python
sns.swarmplot(y='CWDistance',data=df,x='Gender')
plt.title('Swarm Plot Of CWDistance vs. Gender');
```

```python
sns.boxplot(data=df.loc[:,['Age','Height', 'Wingspan', 'CWDistance']])
plt.xticks(fontsize=17,rotation=90)
;

```

```python
sns.boxplot(data=df.loc[df['Gender']=='F',['Age','Height', 'Wingspan', 'CWDistance']])
plt.xticks(fontsize=17,rotation=90)
;
```

```python
px.box(df,y=['Age','Height', 'Wingspan', 'CWDistance','Score'],facet_col='Gender',color='Gender')
```

```python
fig=plt.figure()
ax1=fig.add_subplot(1,3,1)# 1 row 2 columns,position 1
sns.boxplot(data=df.loc[df['Gender']=='M',['Score']])
plt.xticks(fontsize=17,rotation=90)
ax1.title.set_text('Scores M')
ax2=fig.add_subplot(1,3,3)# 1 row 2 columns,position 2
sns.boxplot(data=df.loc[df['Gender']=='F',['Score']])
plt.xticks(fontsize=17,rotation=90)
ax2.title.set_text('Scores F')
;
```
